```python
from rich.tree import Tree
from rich import print as rprint
import os

def create_tree(base_path, ignore_list=None):
    if ignore_list is None:
        ignore_list = []

    base_name = os.path.basename(base_path)
    tree = Tree(base_name)
    for root, dirs, files in os.walk(base_path):
        branch = tree
        parts = root[len(base_path):].split(os.sep)
        for part in parts:
            if part:
                branch = next((b for b in branch.children if b.label == part), branch)
        for file in files:
            if file not in ignore_list:
                branch.add(file)
        dirs[:] = [d for d in dirs if d not in ignore_list]
        for dir in dirs:
            if dir not in ignore_list:
                branch.add(dir)
    return tree

# 设置项目路径和要忽略的文件列表
project_path = "../"
ignore_list = [".gitignore", "requirements.txt", ".git", ".pricingvenv"]

# 创建并打印项目结构树
project_tree = create_tree(project_path, ignore_list)
rprint(project_tree)
```
